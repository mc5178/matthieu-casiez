# Analyse de jeu de données : rendement agricole de la pomme de terre dans 6 pays européens (source : FAOSTAT)

```python
import pandas as pd
data = pd.read_csv("FAOSTAT_data_fr_1-26-2025.csv")
```

Le jeu de données provient de **FAOSTAT**.

FAOSTAT fournit un accès gratuit aux statistiques sur l'alimentation et l'agriculture (y compris les sous-secteurs de la culture, de l'élevage, et de la foresterie) de plus de 245 pays et territoires et couvre tous les groupes régionaux de l'Organisation des Nations unies pour l'alimentation et l'agriculture depuis 1961 jusqu'à l'année disponible la plus récente.

Aperçu des données

    Période : Les données couvrent les années 1961 à 1963.
    Pays inclus : Italie, Pays-Bas, Espagne, Royaume-Uni, France, Allemagne.
    Produit : Les pommes de terre.
    Valeurs observées : Les rendements varient entre environ 9 453 kg/ha (Italie, 1962) et 30 497 kg/ha (Pays-Bas, 1962), avec des fluctuations annuelles.

Observations

    Les Pays-Bas montrent systématiquement les rendements les plus élevés pour les pommes de terre parmi les pays étudiés.
    L'Italie présente des rendements généralement inférieurs à ceux des autres pays européens mentionnés.
    Les données sont qualifiées comme étant des "chiffres officiels", ce qui assure une certaine fiabilité.
